# N Queen Attack Detection

```python
def drawBoard(P):
  N = len(P)
  for i in range(1,N+1):
    print("|", end="")
    for j in range(1,N+1):
      if (i,j) in P:
        ind = str(1+ P.index((i,j)))
        print(" "+ind+" |", end= "")
      else:
        print(" * |", end= "")
    print()
```

```python
import math
import random
def generate_positions(N):
    CL = list(range(0,N*N));
    P = []
    for i in range(N):
        p = random.choice(CL)
        CL.remove(p)
        r = math.ceil((p+1)/N)
        c = 1+p%N
        P.append((r,c))
        print('Q{}: {},{}'.format(i+1,r,c))
    return P

generate_positions(4)
```

```python
import random
def generateRandomPosition(N):
    CL = list(range(0,N*N));
    P = []
    for i in range(N):
        p = random.choice(CL)
        CL.remove(p)
        r = math.ceil((p+1)/N)
        c = 1+p%N
        P.append((r,c))
    return P
```

```python
N = 4
positions = generate_positions(N)
drawBoard(positions)
```

```python
# row attack
def rowAttack(P):
  N = len(P)
  count = 0
  for i in range(N):
    for j in range(i+1,N):
      if P[i][0] == P[j][0]:
        print('row attack:', end = " " )
        print(P[i], end = ", ")
        print(P[j])
        count = count + 1
  return count
```

```python
def rowAttackCount(P):
  N = len(P)
  count = 0
  for i in range(N):
    for j in range(i+1,N):
      if P[i][0] == P[j][0]:
        count = count + 1
  return count
```

```python
#column attack
def columnAttack(P):
  N = len(P)
  count = 0
  for i in range(N):
    for j in range(i+1,N):
      if P[i][1] == P[j][1]:
        print('column attack:', end = " " )
        print(P[i], end = ", ")
        print(P[j])
        count = count + 1
  return count
```

```python
def columnAttackCount(P):
  N = len(P)
  count = 0
  for i in range(N):
    for j in range(i+1,N):
      if P[i][1] == P[j][1]:
        count = count + 1
  return count
```

```python
# diagonal atack
def diagonalAttack(P):
  N = len(P)
  count = 0
  for i in range(N):
    for j in range(i+1,N):
      if abs(P[i][0] - P[j][0]) == abs(P[i][1] - P[j][1]):
        print('diagonal attack:', end = " " )
        print(P[i], end = ", ")
        print(P[j])
        count = count + 1
  return count
```

```python
def diagonalAttackCount(P):
  N = len(P)
  count = 0
  for i in range(N):
    for j in range(i+1,N):
      if abs(P[i][0] - P[j][0]) == abs(P[i][1] - P[j][1]):
        count = count + 1
  return count
```

```python
# implement a method to print a state is valid or not
def isValid(P):
  row = rowAttackCount(P)
  col = columnAttackCount(P)
  diagonal = diagonalAttackCount(P)

  if(row==0 and col==0 and diagonal==0):
    return True
  else:
    return False
```

```python
def isValidWithErrors(P):
  row = rowAttack(P)
  col = columnAttack(P)
  diagonal = diagonalAttack(P)

  if(row==0 and col==0 and diagonal==0):
    return True
  else:
    return False
```

```python
# call isValid untill you get valid solution WITHOUT generating the Board
numOfQueens = 4
randomPosition = generateRandomPosition(numOfQueens)
count = 1

while(not isValid(randomPosition)):
  count = count + 1
  randomPosition = generateRandomPosition(numOfQueens)

drawBoard(randomPosition)
print("valid position reached: " + str(randomPosition))
print("After generating: " + str(count) + " new positions")
```

```python
# call isValid untill you get valid solution WITH board generation for visual perception
numOfQueens = 4
randomPosition = generateRandomPosition(numOfQueens)
drawBoard(randomPosition)
count = 1

while(not isValidWithErrors(randomPosition)):
  print("Invalid positions \n")
  count = count + 1
  print("----------------------------------------")
  randomPosition = generateRandomPosition(numOfQueens)
  drawBoard(randomPosition)

print("valid position reached, \nafter generating: " + str(count) + " new positions")
```
